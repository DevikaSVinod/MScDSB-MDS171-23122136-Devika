```python
Items = ["Book","Paper","Pencil","Pen","Eraser","Hightlighter","Corrector","Sticky notes","File","Adhesive tapes"]
Quantity = [5,12,8,10,9,14,6,11,2,4]
Price = [50,10,15,40,35,80,100,55,125,75]

import random

file = open("Lab 06.txt","w+")
for i in range(0,100):
    item_num = random.choice(Items)
    quant_num = random.choice(Quantity)
    price_num = random.choice(Price)
    total = quant_num*price_num
    print(item_num,quant_num,price_num,total)
    file.write(item_num +","  "\t" + str(quant_num) + "," "\t" + str(price_num)+"," "\t" + str(total)+'\n')

file.close()
```
